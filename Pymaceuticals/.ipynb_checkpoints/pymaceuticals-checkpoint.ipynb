```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_to_load_df=pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data_to_load_df=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
drug_clinical_data=pd.merge(mouse_drug_data_to_load_df,clinical_trial_data_to_load_df)

# Display the data table for preview
drug_clinical_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
Mean_Tumor_Volume=pd.DataFrame(drug_clinical_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean())
Mean_Tumor_Volume=Mean_Tumor_Volume.reset_index()
Mean_Tumor_Volume
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame
Standard_Error_of_Tumor_Volumes=drug_clinical_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem()
Standard_Error_of_Tumor_Volume_df=Standard_Error_of_Tumor_Volumes.reset_index()
Standard_Error_of_Tumor_Volume_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked

tumor_err_response_treatment=Mean_Tumor_Volume.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
tumor_err_response_treatment.reset_index()
tumor_err_response_treatment
```

```python
# Generate the Plot (with Error Bars)

plt.figure (figsize=(16,6))
plt.title('Tumor Response to Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel ('Tumor Volume (mm3)')
plt.grid()
plt.rcParams["font.size"]=12

drug_index= ['Capomulin','Infubinol','Ketapril','Placebo']

for drug in drug_index:
    plot_drug= plt.errorbar(tumor_err_response_treatment.index, tumor_err_response_treatment[drug],fmt='^', 
                            linestyle='-', label=drug)
    plt.legend(loc='upper left')
    if drug== drug_index[-1]:        
# Save the Figure
        plt.savefig("DrugTumorResponseToTreatment.png")
# Show the Figure
plt.show()

# Generate the Plot (with scatter plot)
import itertools

colors = itertools.cycle(["r", "b", "g","y"])

plt.figure (figsize=(10,6))
plt.title('Tumor Response to Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel ('Tumor Volume (mm3)')
plt.grid()
plt.rcParams["font.size"]=12


for drug in drug_index:
    plot_drug= plt.scatter(tumor_err_response_treatment.index, tumor_err_response_treatment[drug], label=drug, color=next(colors), s=25, 
                           marker="o")
    plt.legend(loc='upper left')
    if drug== drug_index[-1]:        
# Save the Figure
        plt.savefig("DrugTumorResponseToTreatmentScatter.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
Metastatic_Sites=pd.DataFrame(drug_clinical_data.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean())
Metastatic_Sites.reset_index()                                                                       
                               
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
Standard_Error_of_Metastatic_Sites=drug_clinical_data.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem()
Standard_Error_of_Metastatic_Sites_df=Standard_Error_of_Metastatic_Sites.reset_index()
Standard_Error_of_Metastatic_Sites_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
metastatic_err_response_treatment=Standard_Error_of_Metastatic_Sites_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')
metastatic_err_response_treatment.reset_index()
metastatic_err_response_treatment
```

```python
# Generate the Plot (with Error Bars)
plt.figure (figsize=(12,6))
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration(Days)')
plt.ylabel ('Metastatic Sites')
plt.grid()
plt.rcParams["font.size"]=10

for drug in drug_index:
    plot_drug= plt.errorbar(metastatic_err_response_treatment.index, metastatic_err_response_treatment[drug],fmt='^', 
                            linestyle='--', label=drug)
    plt.legend(loc='upper left')
    if drug== drug_index[-1]:\
# Save the Figure
        plt.savefig("MetastaticSpreadDuringTreatment.png")
# Show the Figure  
plt.show()

# Generate the Plot (with scatter plot)

plt.figure (figsize=(10,6))
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration(Days)')
plt.ylabel ('Metastatic Sites')
plt.grid()
plt.rcParams["font.size"]=10

for drug in drug_index:
    plot_drug= plt.scatter(metastatic_err_response_treatment.index, metastatic_err_response_treatment[drug], label=drug, color=next(colors), s=25, 
                           marker="o",linewidth=4)
    plt.legend(loc='upper left')
    if drug== drug_index[-1]:
# Save the Figure
        plt.savefig("MetastaticSpreadDuringTreatmentScatter.png")
# Show the Figure  
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
# Preview DataFrame
mouse_survival_rate= drug_clinical_data.groupby(['Drug','Timepoint'])['Mouse ID'].count()
mouse_survival_rate_df=mouse_survival_rate.reset_index()
mouse_survival_rate_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview the Data Frame
mouse_survival_rate_to_treatment=mouse_survival_rate_df.pivot(index='Timepoint',columns='Drug',values='Mouse ID')
mouse_survival_rate_to_treatment.reset_index()
mouse_survival_rate_to_treatment
```

```python
# Generate the Plot (Accounting for percentages)

plt.figure (figsize=(12,6))
plt.title('Mouse Survival Rate to Treatment')
plt.xlabel('Treatment Duration(Days)')
plt.ylabel ('Survival Rate(%)')
plt.grid()
plt.rcParams["font.size"]=12

for drug in drug_index:
    plot_drug= plt.errorbar(mouse_survival_rate_to_treatment.index, mouse_survival_rate_to_treatment[drug],fmt='^', 
                            linestyle='--', label=drug)
    plt.legend(loc='lower left')
    if drug== drug_index[-1]:\
# Save the Figure
        plt.savefig("MouseSurvivalRatetoTreatment.png")
# Show the Figure
    
plt.show()

# Generate the scatter Plot (Accounting for percentages)
plt.figure (figsize=(10,6))
plt.title('Mouse Survival Rate to Treatment')
plt.xlabel('Treatment Duration(Days)')
plt.ylabel ('Survival Rate(%)')
#plt.grid()
plt.rcParams["font.size"]=12

for drug in drug_index:
    plot_drug= plt.scatter(mouse_survival_rate_to_treatment.index, mouse_survival_rate_to_treatment[drug], label=drug, color=next(colors), s=25, 
                           marker="o")
    plt.legend(loc='lower left')
    if drug== drug_index[-1]:\
# Save the Figure
        plt.savefig("MouseSurvivalRatetoTreatmentScatter.png")
# Show the Figure
    
plt.show()

```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Display the data to confirm
tumor_volume_pct_change=(tumor_err_response_treatment.iloc[-1]/(tumor_err_response_treatment.iloc[0])-1)*100
tumor_volume_pct_change
```

```python
# Splice the data between passing and failing drugs
passing = tumor_volume_pct_change < 0 

# Orient widths. Add labels, tick marks, etc. 

pct_change_list=[(tumor_volume_pct_change[drug]) for drug in drug_index]
pct_change_plot= plt.bar(drug_index,pct_change_list,width=-1,align='edge',color=passing.map({True:'g', False:'r'}))

plt.grid()
plt.ylim(-30,70)
plt.xlabel('Drug')
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor change over treatment duration')

# Use functions to label the percentages of changes
def setlabel(rects):
    for rect in rects :
        height=rect.get_height()
        if height > 0:
                lebel_position= 4
        else:
                lebel_position= -8
                
        plt.text(rect.get_x()+rect.get_width()/2., lebel_position,'%d' % int(height)+'%', color='white', ha='center', va='bottom')

# Call functions to implement the function calls
setlabel(pct_change_plot)

# Save the Figure
plt.savefig("TumorVolumePctChangeOvertime.png")

# Show the Figure
plt.show()
```
